## Linear Regression

Given input data $X$ and $Y$, a simple linear regression model: 
$\hat{Y}=Xw + b + \epsilon$, where we have:


**Data (our inputs)**:  
- $X$: The input data.
- $Y$: The target data we want to fit with.
- $\hat{Y}$: Your predicted target

**Model parameter**: 
- $w$: The weight term of the linear model
- $b$: Bias term of the linear model

**Noise**: 
- $\epsilon$ is the noise term -- we always have noise that we don't know.



# Lets use Pytorch to solve it!

## Make synthetic data for Regression
https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html

Generating Simple Linear Regression Model:
$\hat{Y}=Xw + b + \epsilon$, where
- b: bias term can be specified by `bias`. By default, $b=0$
- $\epsilon$ is the Gaussian noise $\sim \text{N}(0, \sigma^2)$. $\sigma$ can be specified by `noise`. By default, $\sigma=0$.

`make_regression` returns:
- X: input samples (`n_samples`, `n_features`)
- y: target data (`n_samples`, `n_targets`). `n_targets` is 1 by default.
- w: weights (`n_features`, `n_targets`) when `coef` is set to be $\text{True}$. `n_targets` is 1 by default.

```python
n_samples = 100
n_features = 1
n_informative = 1 # number of features used to generate y
noise = 10
def make_regression(n_samples, n_features, n_informative, noise):
    X, y, w = datasets.make_regression(n_samples=n_samples, n_features=n_features,
                                      n_informative=n_informative, noise=noise, bias = 0, coef=True, random_state=0)
    return X, y, w
X, y, w = make_regression(n_samples, n_features, n_informative, noise)
print(X.shape,y.shape,w.shape)
plt.scatter(X[:,0], y, color='r', marker='.')
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```

## Two ways to solve it: 

### 1) Closed form solution

Suppose that $X \in \mathbb{R}^{n \times m}$ with $X^TX$ invertible (nonsigular) and $Y \in \mathbb{R}^n$, and that $Y \sim N(Xw, \sigma^2 I)$.  
You learned in class that the maximum likelihood estimate $\hat{w}$ of ${w}$ is given by

$\hat{w} = (X^TX)^{-1}X^TY$


### Why? 
The loss function $L(w) = (Y-\hat{Y})^2$ 

If we expand it...

$L(w) = (Y-\hat{Y})^T(Y-\hat{Y}) = (Y-Xw)^T(Y-Xw) = Y^TY-Y^TXw-w^TX^TY+(Xw)^TXw$  

$= Y^TY-2w^TX^TY+w^TX^TXw$  

$\frac{\partial L(w)}{\partial w} = 0 - 2X^TY + 2X^TXw$  


Set $\frac{\partial L(w)}{\partial w} = 0$ we have  


$X^TXw = X^TY \Rightarrow \hat{w}=(X^TX)^{-1}X^TY$

```python
X, y, w = torch.from_numpy(X).float(), torch.from_numpy(y).float(), torch.from_numpy(w).float()

## get w using closed-form solution
w_hat = X.t().matmul(X).inverse().matmul(X.t()).matmul(y)
y_hat = X.matmul(w_hat)

## compare the predicted y_hat with the ground truth y
criterion = nn.MSELoss()
mse = criterion(y_hat,y) # (y_hat-y).pow(2).mean()
print('True parameters:', w)
print('Estimated parameters by closed-form solution:', w_hat)
print('Mean Squared Error (MSE):', mse.item())
plt.scatter(X[:,0].numpy(), y.numpy(), color='r', marker='.')
plt.plot(X[:,0].numpy(), y_hat.numpy(), color='blue', linewidth=3, label='Closed form')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```

### 2) Optimization

We use **Gradient Descent** to optimize the linear regression model $Y=Xw$.

```python
num_iter = 50  ## Number of iterations
w_true = torch.unsqueeze(w, 0)
w_hat = torch.zeros(n_features).requires_grad_(True)
lr = 0.1
# criterion = nn.MSELoss()
def mse(preds, targets):
    return ((preds-targets)**2).mean()
def model(X, w):
    return torch.matmul(X,w)

losses = []
for i in range(num_iter):
    y_hat = model(X, w_hat)
    loss = mse(y_hat, y)
#     loss = criterion(y_hat, y)
    loss.backward()
    with torch.no_grad():
        w_hat.data -= lr * w_hat.grad
#         print(f'Before zero out {w_hat.grad}')
        # zero out gradients
        w_hat.grad.zero_()
#         print(f'After zero out {w_hat.grad}')
    losses.append(loss.item())
#     print(loss)

print('True parameters: ', w_true)
print('Estimated parameters by GD: ', w_hat.data)
print('Mean Squared Error (MSE): ', loss.data)
plt.figure(figsize=(8,4))
plt.scatter(X[:,0].numpy(), y.numpy(), color='r', marker='.')
plt.plot(X[:,0].numpy(), y_hat.detach().numpy(), color='blue', linewidth=2, label='Gradient Descent')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(list(range(len(losses))), losses, color='blue', linewidth=2, label='Learning curve')
plt.legend()
plt.xlabel('iterations')
plt.ylabel('MSE')
plt.grid()
plt.show()
```

### Closed form solution vs. Optimization
You may ask why we need optimization method as we can always derive the closed form solutions.

Recall closed form solution: $\hat{w}=(X^TX)^{-1}X^TY$.

The [complexity](https://en.wikipedia.org/wiki/Computational_complexity_of_mathematical_operations) (in general case and without any structure on $M$) of inverting a matrix $M \in R^{n \times n}$ is $O(n^3)$. Imagine we have a model with a large number of parameters, inverting a matrix actually is much more computationally expensive than gradually optimizing your model.

Below computes the time cost of each approach for solving a linear regression problem w.r.t. different number of input features.

**Not every model has a closed form solution!**

```python
def mse(preds, targets):
    return ((preds-targets)**2).mean()
def model(X, w):
    return X@w

n_samples = 100000
n_features = [100, 500, 1000, 1500, 3000, 4500]
n_informative = n_features
noise = 10
criterion = nn.MSELoss()
closed_form_time = []
closed_form_mse = []
optimization_time = []
optimization_mse = []
for i in range(len(n_features)):
    # make synthetic data for different numbers of features
    X, y, w = make_regression(n_samples, n_features[i], n_informative[i], noise)
    X, y ,w = torch.from_numpy(X).float(), torch.from_numpy(y).float(), torch.from_numpy(w).float()
    s = time.time()
    
    #Closed form
    w_hat = X.t().matmul(X).inverse().matmul(X.t()).matmul(y)    
    y_hat = X.matmul(w_hat)
    closed_form_time.append(time.time()-s)
    closed_form_mse.append(criterion(y_hat,y))
    
    #optimization
    s = time.time()
    num_iter = 100  ## Number of iterations
    w_true = torch.unsqueeze(w, 0)
    w_hat = torch.zeros(n_features[i]).requires_grad_()
    lr = 0.1
    losses = []
    max_loss = 1e10

    for i in range(num_iter):
        y_hat = model(X, w_hat)
        loss = mse(y_hat, y)
        loss.backward()
        with torch.no_grad():
            w_hat.data -= lr * w_hat.grad
            # zero out gradients
            w_hat.grad.zero_()
        # record best model
        if loss < max_loss:
            max_loss = loss
            w_best = w_hat
            t_best = time.time() 
    y_hat = model(X, w_best)
    optimization_time.append(time.time()-t_best)
    optimization_mse.append(mse(y_hat, y))
closed_form_time, closed_form_mse = np.array(closed_form_time, dtype=np.float32), np.array(closed_form_mse, dtype=np.float32)
optimization_time, optimization_mse = np.array(optimization_time, dtype=np.float32), np.array(optimization_mse, dtype=np.float32)

plt.figure(figsize=(10,10))
plt.plot(closed_form_time, closed_form_mse, color='red', linewidth=3, label='Closed Form')
plt.plot(optimization_time, optimization_mse, color='blue', linewidth=3, label='Optimization')
plt.scatter(closed_form_time, closed_form_mse, color='red')
plt.scatter(optimization_time, optimization_mse, color='blue')
for i, txt in enumerate(n_features):
    plt.annotate(str(txt), (closed_form_time[i], closed_form_mse[i]))
    plt.annotate(str(txt), (optimization_time[i], optimization_mse[i]))
plt.legend()
plt.xlabel('Time(s)')
plt.ylabel('MSE')
plt.grid()
plt.show()
```

## [Logistic Regression](https://en.wikipedia.org/wiki/Logistic_regression)

Simple Logistic Regression Model

$Y = Sigmoid(Xw) = \frac{1}{1+exp(Xw)}$, for two-class classification.

$Y = Softmax(Xw) = \frac{exp(Xw_i)}{\sum_{j}exp(Xw_j)}$,for multi-class classification.  
where we have:

**Data (our inputs)**:  
- X: The input data.
- Y: The target data we want to fit with.

**Model parameter**: 
- w: The weight term of the linear model


## Make synthetic data for Classification
https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html#sklearn.datasets.make_classification

Generate a random n-class classification problem.

`make_classification` returns
- X: Generated samples. (n_samples, n_features)
- y: The integer labels. (n_samples,)

```python
n_samples = 1000
n_features = 2
n_informative = 2 # number of features used to generate y
n_classes = 4
def make_classification(n_samples, n_features, n_informative, n_classes):
    X, y = datasets.make_classification(n_samples=n_samples, n_features=n_features,
                                      n_informative=n_informative, n_redundant=0, n_classes=n_classes, 
                                       n_clusters_per_class=1, random_state=0)
    return X, y
X, y = make_classification(n_samples, n_features, n_informative, n_classes)
print(X.shape,y.shape)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25)
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```

Logistic Regression generally does not have a closed form solution. We usually use SGD (training with minibatches) to fit the model iteratively. Here we use all data to train the model (simply gradient descent) since we do not have the problem of memory limitation, and thus not stochastic.

Note: `nn.CrossEntropyLoss()` combines `LogSoftmax` and `NLLLoss` in one single class.

```python
#### Show all your trainable weights in the model
# for name, param in model.named_parameters():
#     if param.requires_grad:
#         print(name)
#         print('Data', param.data)
#         print('Gradient', param.grad)
```

```python
X, y = torch.from_numpy(X).float(), torch.from_numpy(y).long()

num_iter = 30

## this time we use torch.nn to define the model
model = nn.Linear(in_features=n_features, out_features=n_classes, bias=False) # y = Xw

optimizer = torch.optim.SGD(model.parameters(), lr=1) # Experiment to see the senstivity of learning rate with SGD
criterion = nn.CrossEntropyLoss()
# Train
model.train()
losses = []
for i in range(num_iter):
    y_hat = model(X) # Forward Pass
    loss = criterion(y_hat, y) # -log(softmax(y_hat)) == nn.NLLLoss(nn.LogSoftmax(y_hat)) == nn.CrossEntropyLoss(y_hat)
    optimizer.zero_grad()
    loss.backward() # Backward Pass
    losses.append(loss.item())
    optimizer.step() # we are using a optimizer now!
# Evaluate
with torch.no_grad():
    model.eval()
    y_hat = model(X)
    nll = criterion(y_hat.squeeze(),y)
print('Estimated parameters: ', model.weight)
print('Negative Log Likelihood (NLL): ', nll.item())
```

```python
plt.figure()
plt.plot(list(range(len(losses))), losses, color='blue', linewidth=3, label='Learning curve')
plt.legend()
plt.xlabel('iterations')
plt.ylabel('NLL')
plt.grid()
plt.show()
```

```python
def plot_decision_boundary(model, X):
    # Set min and max values and give it some padding
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = torch.meshgrid(torch.arange(x_min, x_max, h), torch.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    X = torch.cat([xx.reshape(-1,1), yy.reshape(-1,1)],dim=1)
    y_hat = model(X)
#     print('y_hat:', y_hat)
    Z = y_hat.topk(1,1,True,True)[1]
#     print('Z:' , Z)
    Z = Z.view(xx.size())
#     print('reshaped Z: ', Z)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    return
    
    
plt.figure()
plot_decision_boundary(model, X)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=25)
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```
